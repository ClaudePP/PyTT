```python
from Modules import MaterialBank as mb
from Modules import NecessaryVariables as nv
import numpy as np

from matplotlib import pyplot as plt

%matplotlib inline
```

```python
mat1 = mb.Material("MaterialInfo/Tungsten.txt")
mat2 = mb.Material("MaterialInfo/Molybdenum.txt")
```

```python
print("material 1 density [g/cm3] = ",mat1.rho)
print("material 2 density [g/cm3] = ",mat2.rho)
```

```python
#temp=[[300,1000,2200]]
temp= np.array([np.array(np.arange(300,2300,100))])
print("temperatures = ",temp)
print(mat2.GetEmissivity(temp))
emis1=mat1.GetEmissivity(temp)
emis2=mat2.GetEmissivity(temp)
    
```

```python
fig, ax = plt.subplots(1, figsize=(7.5,6))
ax.plot(temp[0],emis1[0],label="tungsten")
ax.plot(temp[0],emis2[0],label="molybdenum")
ax.set_xlabel('temperature [K]',fontsize=12)
ax.set_ylabel('emissivity',fontsize=12)
#ax.set_xlabels(fontsize=12)
ax.grid()
ax.legend(fontsize=12)

plt.savefig("figures/emittances.png")
```
