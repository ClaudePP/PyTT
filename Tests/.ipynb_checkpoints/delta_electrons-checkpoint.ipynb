# delta electron cross section

```python
from Modules import DeltaElectrons as de
from Modules import RelativisticFormulas as re
from Modules import NecessaryVariables as nv

from matplotlib import pyplot as plt
import numpy as np
import scipy.constants as con
from scipy.integrate import quad
```

```python
ened = np.logspace(0,6,50)  # electron energies in eV (1 eV  -  1 MeV)
#ened = np.linspace(1,100000,50)  # electron energies in eV

#enedJ = ened/nv.kgMeV*1e6
#enedJ = ened*nv.Qe  # electron energies in Joules
```

```python
# projectile energy
Ep=100e6   # [eV]
Zt=6       # target nucleai charge
Zeff=1     # projectile effective Z

beta,gamma = re.beta_gamma_from_energy(Ep/1e6,con.physical_constants['proton mass energy equivalent in MeV'][0])

print("relativistic beta = ",beta)
#Eshell=99.42  # [eV] # free silicon
#Eshell=33.6  # [eV]  # M-shell, Silicon bulk
Eshell=4.5    # [eV] graphite has electrons in conduction band, so I use only work function here, is it correct?

print("maximum energy transfer [J] =",de.Edmax(beta,Eshell))
print("maximum energy transfer [MeV] =",de.Edmax(beta,Eshell)/(1e6*nv.Qe))

# delta electron differential cross section
# ened - in eV
decs = de.DiffECrossSection(ened,Zt,beta,Zeff,Eshell) # [m^2/J]


#print(con.eV)
# convert to cm^2/eV, to compare with papers:
# 1 m2 = 100 cm2
# 1 eV = 1.602e-19 J -> 1J = 1/1.602e-19 eV
# 1 m2/J = 100^2 cm2/(1/1.602e-19) eV = 100^2*1.602e-19 cm2/eV
decs_cm2eV = decs*pow(100,2)*con.eV
```

```python
fig, ax = plt.subplots(1, figsize=(7.5,6))
ax.plot(ened,decs_cm2eV/1e-16,label="differential cross section")
ax.set_xlabel("electron energy [eV]",fontsize=12)
ax.set_ylabel("differential cross section [$10^{-16} cm^2/eV$]",fontsize=12)
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1.0,2e5)
ax.set_ylim(5.e-13,1e-1)
ax.legend(fontsize=12)
ax.grid()
```

```python
# electron range in carbon/graphite
At=12.0
rho=2.0  # [g/cm3]
# electron range in material
# ened in eV
# rho in g/cm3
eranges = de.PenetrationDepth(ened,Zt,At,rho) # [m]
```

```python
fig, ax = plt.subplots(1, figsize=(7.5,6))
ax.plot(ened,eranges,label="electron range in graphite")
ax.set_xlabel("electron energy [eV]",fontsize=12)
ax.set_ylabel("electron range [m]",fontsize=12)
plt.hlines(15e-6,ened.min(),ened.max(), color='red',linestyles='dashdot')
plt.vlines(4e4,eranges.min(),eranges.max(), color='red',linestyles='dashdot')

axcs = ax.twinx()
axcs.plot(ened,decs_cm2eV/1e-16,'g',label="differential cross section")
axcs.set_yscale('log')
axcs.set_ylabel("differential cross section [$10^{-16} cm^2/eV$]",fontsize=12)

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlim(1.0,1e6)
ax.set_ylim(5.e-13,1e-1)
ax.legend(fontsize=12)
ax.grid()
```

```python
I = quad(de.DiffECrossSection,4e4,1e6, args=(Zt,beta,Zeff))  # [m2/J]
#I = quad(decs,4e4*nv.Qe,1e6*nv.Qe)
print("total cross section [m2]: ")
print(I)
# current: - makes no sense
#print("total current [A]:")
#print(I[0]*nv.Qe)
```

```python
# SPS WS 2008 parameters
Ntot=2e13     # number of circulating protons
dwcm=30e-6       # [m]
fsps=43478.26 # [Hz]
vwcm=1.0        # [m/s]
Mt=12.01e-3        # kg/mole
rho=2*0.001/pow(0.01,3)     # kg/m3

# well-known formula, see paper from Quench test:
Npass=Ntot*fsps*dwcm/vwcm


Ndelta=Ntot*rho*nv.Na*fsps*dwcm**2*I[0]*nv.Qe/(4*Mt*vwcm)  # this seems to make sense, but is it correct?
# this should be verified with Geant4

nfact=rho*nv.Na*dwcm*I[0]/(4*Mt)


print("number of passing protons/1e10: ",Npass/1e10)
print("number of generated delta electrons/1e10: ",Ndelta/1e10)
print(nfact)
print("total probability of delta electron emission: ",Ndelta/Npass)
print("total energy removed [keV]: ",Ndelta*50/Npass)
```
