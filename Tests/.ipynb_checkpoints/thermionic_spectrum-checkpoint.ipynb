plot spectrum of thermionic electrons

```python
from Modules import NecessaryVariables as nv
from Modules import TempPhysicalModels as tpm
from matplotlib import pyplot as plt
import numpy as np
```

```python
enes = np.logspace(-4,1.2,100)  # energies in eV
vels = [tpm.ElectronVelocity(Eks)/1000 for Eks in enes]
```

```python
spect2000 = tpm.ThermionicElectronSpectra(2000,enes)
spect2500 = tpm.ThermionicElectronSpectra(2500,enes)
spect3000 = tpm.ThermionicElectronSpectra(3000,enes)
spect3500 = tpm.ThermionicElectronSpectra(3500,enes)
```

```python
print(enes)
print(spect3500)
```

```python
fig, ax = plt.subplots(1, figsize=(7.5,6))
ax.plot(enes,spect2000,label="2000 K")
ax.plot(enes,spect2500,label="2500 K")
ax.plot(enes,spect3000,label="3000 K")
ax.plot(enes,spect3500,label="3500 K")
#plt.yscale('log')
axv = ax.twiny()
axv.set_xlabel("velocity [km/s]", fontsize=12)
axv.plot(vels,spect2000)
axv.set_xscale('log')

ax.set_xscale('log')
#plt.yticks([])
plt.yticks(fontsize=12, color="w")
ax.set_xlabel('kinetic energy [eV]',fontsize=12)
ax.set_ylabel('current [a.u.]',fontsize=12)
#ax.set_xlabels(fontsize=12)
ax.grid()
ax.legend(fontsize=12)
```
