```python
import matplotlib.pyplot as plt
import pandas as pd

from platform import python_version
print("python version: ",python_version())
#matplotlib notebook
%matplotlib inline
```

```python
#df = pd.read_csv("Last.txt", skiprows=9)
#df = pd.read_csv("Last_TATTOOS_Mo100um_100uA.txt", skiprows=9)
#df = pd.read_csv("Last_TATTOOS_W100um_100uA.txt", skiprows=9)
df = pd.read_csv("Last_UCN_harp_Mo40um_1p9mA.txt", skiprows=9)
```

```python
print(df)
```

```python
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["MaxTemp1[K]"][:-1],label="central wire")
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["MaxTemp0[K]"][:-1],label="side wire")
plt.xlabel("Time [s]")
plt.ylabel("Maximum Wire Temperature [K]")
plt.legend()
plt.text(1.5,850,r"wire: Mo, $\rm 40 \mu m$")
#plt.text(1.5,750,r"beam current: $\rm 100 \mu A$")
plt.text(1.5,750,r"beam current: 1.9 mA")
plt.text(1.5,650,r"beam sigma: 4.6 mm")
plt.grid()
plt.savefig("UCN_harp_temp_evol")
```

```python
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["SEMcurr1[uA]"][:-1],label="central wire")
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["SEMcurr0[uA]"][:-1],label="side wire")
plt.xlabel("Time [s]")
plt.ylabel("Secondary Electron Current [uA]")
plt.legend()
```
