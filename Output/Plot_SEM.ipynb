```python
import matplotlib.pyplot as plt
import pandas as pd

from platform import python_version

print("python version: ",python_version())
```

```python
df = pd.read_csv("Last.txt", skiprows=9)
```

```python
print(df)
```

```python
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["MaxTemp1[K]"][:-1],label="central wire")
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["MaxTemp0[K]"][:-1],label="side wire")
plt.xlabel("Time [s]")
plt.ylabel("Maximum Wire Temperature [K]")
plt.legend()
```

```python
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["SEMcurr1[uA]"][:-1],label="central wire")
plt.plot([x*1e-6 for x in df["Time[us]"][:-1]],df["SEMcurr0[uA]"][:-1],label="side wire")
plt.xlabel("Time [s]")
plt.ylabel("Secondary Electron Current [uA]")
plt.legend()
```
