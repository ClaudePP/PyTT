```python
filein = open('IntensityVSTime_Emi.txt', 'r')
wpos=[]
wsig=[]
time=[]
signal=[]
for lnr,lne in enumerate(filein):
    if lnr<4:
        print(lnr,lne)
    if lnr>3:
        tme,wpos,wsig=lne.split("   ")
        time.append(float(tme))
        signal.append(float(wsig[3:-6]))
        if lnr%100==1:
            print(tme,",",float(wsig[3:-6]))
```

```python
plt.plot(time,signal)
plt.xlabel("Time [us]")
plt.ylabel("Current [A]")
plt.xlim(0.9995e6,1.001e6)
```
