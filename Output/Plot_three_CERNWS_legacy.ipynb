```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
import numpy as np
```

```python
file1="CERN_SPS_BWS_2008_01mps.txt"
file2="CERN_SPS_BWS_2008_1mps_b.txt"
file3="CERN_SPS_BWS_2008_02mps.txt"

# read simulation output files
with open(file1) as f:
    cfname=f.readlines()[1]
confname1=cfname[26:]    
print(confname1)    


# read simulation output files
with open(file2) as f:
    cfname=f.readlines()[1]
confname2=cfname[26:]    
print(confname2)    
```

```python
df1 = pd.read_csv(file1, skiprows=9)
df2 = pd.read_csv(file2, skiprows=9)
df3 = pd.read_csv(file3, skiprows=9)
```

```python
# rename columns
df1.rename({"#   Time [us]":"time"," Position [mm]":"pos"," Npart":"npart"," MaxTemp [K]":"maxt"," SEMcurr [uA]":"isem"," THcurr [uA] ":"ithc"}, axis='columns',inplace=True)
df2.rename({"#   Time [us]":"time"," Position [mm]":"pos"," Npart":"npart"," MaxTemp [K]":"maxt"," SEMcurr [uA]":"isem"," THcurr [uA] ":"ithc"}, axis='columns',inplace=True)
df3.rename({"#   Time [us]":"time"," Position [mm]":"pos"," Npart":"npart"," MaxTemp [K]":"maxt"," SEMcurr [uA]":"isem"," THcurr [uA] ":"ithc"}, axis='columns',inplace=True)
```

```python
fig, ax = plt.subplots(1, figsize=(7.5,6))

hsimt1, = ax.plot([x/1e6 for x in df1["time"]],df1["maxt"],'g--',label="0.1 m/s")
hsimt3, = ax.plot([x/1e6 for x in df3["time"]],df3["maxt"],'b--',label="0.2 m/s")
hsimt2, = ax.plot([x/1e6 for x in df2["time"]],df2["maxt"],'r--',label="1 m/s")
ax.set_xlabel("time [s]", fontsize=12)
ax.grid()
ax.set_ylabel("max temperature [K]", fontsize=12)
ax.legend(fontsize=12)
plt.savefig('three.png')
```
